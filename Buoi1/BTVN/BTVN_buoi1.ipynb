**Câu 3: Python là ngôn ngữ biên dịch hay thông dịch? Vì sao?**

- Python là ngôn ngữ thông dịch. Điều này có nghĩa là mã Python được thực thi bởi một trình thông dịch (interpreter) từng dòng một thay vì được biên dịch toàn bộ thành mã máy trước khi chạy (như trong ngôn ngữ biên dịch).

- Lý do:

+ Python sử dụng trình thông dịch (ví dụ như CPython) để chuyển đổi mã nguồn thành mã bytecode. Trình thông dịch sau đó thực thi mã bytecode này. Do đó, Python không cần phải biên dịch toàn bộ trước khi chạy mà thực thi mã trực tiếp.
+ Việc sử dụng trình thông dịch làm cho Python linh hoạt hơn và dễ phát triển, nhưng cũng khiến Python chậm hơn so với các ngôn ngữ biên dịch như C++

Câu 5:
1. Cách khai báo biến: Trong python không cần khai báo kiểu dữ liệu cho biến
VD: a=100, name= "Huy" , b=1.25
2. Các toán tử trong Python: Toán tử số học: +, -, *, /, // (chia lấy nguyên), % (chia lấy dư), ** (Phép lấy số mũ); Toán tử so sánh: ==, !=, >, <, >=, <= ;Toán tử logic: and, or, not; Toán tử gán: =, +=, -=, *=, /=, ...; Toán tử bit: &, |, ^, ~, <<, >>
3. Các kiểu dữ liệu trong Python
Số nguyên (integer): là một loại dữ liệu số vô hướng, ví dụ như 1, 2, 3, 4, 5.
Số thực (float): là một loại dữ liệu số có dấu thập phân, ví dụ như 1.5, 2.7, 3.14.
Chuỗi (string): là một loại dữ liệu chứa một chuỗi ký tự, ví dụ như "hello", "goodbye".
Boolean: là một loại dữ liệu có hai giá trị là True hoặc False.
Danh sách (list): là một loại dữ liệu chứa một tập hợp các giá trị, ví dụ như [1, 2, 3, 4, 5].
Từ điển (dictionary): là một loại dữ liệu chứa các cặp khóa-giá trị, ví dụ như {"name": "John", "age": 30}.


```python
#Ví dụ về khai báo biến
age = 19         # Biến nguyên
name = "Huy"   # Biến chuỗi
b = True  # Biến boolean
```

```python
#Toán tử số học
a = 5
b = 3
print(a + b)   # 8
print(a - b)   # 2
print(a * b)   # 15
print(a / b)   # 1.666...7
print(a // b)  # 1 (chia lấy nguyên)
print(a % b)   # 2 (chia lấy dư)
print(a ** b)  # 125 (5 mũ 3)
```

```python
#Toán tử so sánh
x = 2
y = 1
print(x == y)   # False
print(x != y)   # True
print(x > y)    # True
print(x < y)    # False
print(x >= y)   # True
print(x <= y)   # False
```

```python
#Toán tử logic
x = True
y = False
print(x and y)   # False
print(x or y)    # True
print(not x)     # False
```

```python
#Toán tử gán
x = 5
x += 3    # x = x + 3 -> 8
x -= 2    # x = x - 2 -> 8-2=6
x *= 4    # x = x * 4 -> 6x4=24
x /= 6    # x = x / 6 -> 24/6=4
print(x)
```

```python
#Toán tử bit
a = 60    # 60 = 0011 1100
b = 13    # 13 = 0000 1101
print(a & b)   # 12 (0000 1100)
print(a | b)   # 61 (0011 1101)
print(a ^ b)   # 49 (0011 0001)
print(~a)      # -61 (bit đảo ngược của a)
print(a << 2)  # 240 (dịch trái 2 bit, 0011 1100 -> 1111 0000)
print(a >> 2)  # 15 (dịch phải 2 bit, 0011 1100 -> 0000 1111)
```

```python
#Ví dụ về các kiểu dữ liệu

# Số nguyên (int)
age = 19
print("age:", age, "| type:", type(age))

# Số thực (float)
height = 1.70
print("height:", height, "| type:", type(height))

# Chuỗi (str)
name = "Huy"
print("name:", name, "| type:", type(name))

# Boolean (bool)
is_student = True
print("is_student:", is_student, "| type:", type(is_student))

# Danh sách (list)
fruits = ["apple", "banana", "mango"]
print("fruits:", fruits, "| type:", type(fruits))

# Tuple
my_tuple = (10, 20)
print("my_tuple:", my_tuple, "| type:", type(my_tuple))

# Từ điển (dict)
person = {"name": "Huy", "age": 19}
print("person:", person, "| type:", type(person))

# Set
my_numbers = {1, 2, 3, 4, 5}
print("my_numbers:", my_numbers, "| type:", type(my_numbers))
```
